```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
```

# Функция одной переменной

```python
x = np.arange(-10, 10.01, 0.01)
plt.plot(x, 3 * x ** 2)
plt.show()
```

```python
x0 = 5
l = 0.1
res = []
```

```python
def norm(vector):
    return np.linalg.norm(vector)
```

```python
def f(x):
    return 3 * x ** 2

def df(x):
    return 6 * x

while True:
    res.append([x0, f(x0)])
    x1 = x0 - l * df(x0)
    if norm(x0 - x1) <= 0.00000001:
        break
    x0 = x1
print(res)
```

```python
res = np.array(res)
res
```

```python
plt.plot(res[:, 0], res[:, 1], marker='o', color='red')
x = np.arange(-5, 5, 0.01)
plt.plot(x, 3 * x ** 2)
plt.show()
```

# Функция двух переменных

```python
def f(x, y):
    return 3*x**2 + 5*y**2
def df(x, y):
    return np.array([6*x, 10*y])
```

```python
x0 = np.array([5,0])
l = 0.1
res = []
```

```python
while True:
    res.append(x0.tolist())
    x1 = x0 - l * df(*x0)
    if norm(x1 - x0) <= 0.00000001:
        break
    x0 = x1
print(res)
```

```python
res = np.array(res)
res
```

```python
x = np.linspace(-2, 2, 100)
y = np.linspace(-2, 2, 100)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z)

path = np.array(res)
ax.plot(path[:, 0], path[:, 1], f(path[:, 0], path[:, 1]), color='r', marker='o')

ax.scatter(x0[0], x0[1], f(*x0), color='red', s=100)

plt.legend()
plt.show()
```
